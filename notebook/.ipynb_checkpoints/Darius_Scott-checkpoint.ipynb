```python
def get_train_data(n1, n2):
    x_train = []
    y_train = []
    number = 0
    count = 0
    indices = []
    x = True
    while x == True:
        row = data.loc[number][6]
        #print(row)
        if(row == "NO"):
            x_train.append(data.loc[number][:6])
            y_train.append(data.loc[number][6])
            count += 1
            number += 1
            indices.append(number)
            if(count == n1):
                x = False
        else:
            number += 1

    number = 0
    count = 0
    x = True
    while x == True:
        row = data.loc[number][6]
        #print(row)
        if(row == "AB"):
            x_train.append(data.loc[number][:6])
            y_train.append(data.loc[number][6])
            count += 1
            number += 1
            indices.append(number)
            if(count == n2):
                x = False
        else:
            number += 1
    x_train = np.array(x_train)
    y_train = np.array(y_train)
    return x_train, y_train, indices
```

```python
def get_test_data(indices):
    x_test = []
    y_test = []
    for i in range(len(data)):
        if i not in indices:
            x_test.append(data.loc[i][:6])
            #print(i)
            #print(data.loc[number][:6])
            y_test.append(data.loc[i][6])
            i += 1
    x_test = np.array(x_test)
    y_test = np.array(y_test)  
    return x_test, y_test
```

```python
def convert_numpy(x1,x2,y1,y2):
    x_train = np.array(x1)
    y_train = np.array(y1)
    x_test = np.array(x2)
    y_test = np.array(y2)
```

```python
def convert_classifier(input_val):
    encoder = preprocessing.LabelEncoder()
    output = encoder.fit_transform(input_val)
    return output
```

### c. Write code for k-nearest neighbors with Euclidean metric (or use a software package).

```python
def KNN_model(n, m, w):
    model = KNeighborsClassifier(n_neighbors=n, metric = m, weights = w)
    return model
```

```python
def get_k_value_scores(x_train, y_train, x_test, y_test, increment, input_metric, input_weights):
    k_values = []
    train_errors = []
    test_errors = []
    k = 1
    for x in range(len(x_train)):
        if(k < x): #number of neighbors has to be less than the size of training set
            model = KNN_model(k, input_metric, input_weights)
            model.fit(x_train, y_train)
            train_errors.append(1 - model.score(x_train, y_train))
            test_errors.append(1 - model.score(x_test,y_test))
            #pred_test = model.predict(x_test)
            #accuracy = metrics.accuracy_score(pred, y_test)
            #test_errors.append(1 - accuracy)
            k_values.append(k)
            k += increment
    return k_values, train_errors, test_errors
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_score
from sklearn.metrics import f1_score
def get_data_accuracy(y_test, pred):
    true_negative, false_positive, false_negative, true_positive  = confusion_matrix(y_test,pred).ravel()
    precision = precision_score(y_test, pred)
    f1 = f1_score(y_test, pred)
    return true_negative, false_positive, false_negative, true_positive, precision, f1
```

```python
def get_best_test_error(input_metric, input_weights):
    boolean = True
    N = 10
    train_list = []
    error = []
    k_values = []
    while boolean == True:
        x_train, y_train, indices = get_train_data(int(round(N/3,0)), int(round((N - (N/3)),0)))
        x_test, y_test = get_test_data(indices)
        y_train = convert_classifier(y_train)
        y_test = convert_classifier(y_test)
        k_values, train_scores, test_scores = get_k_value_scores(x_train, y_train, x_test, y_test,
                                              increment=5,
                                              input_metric=input_metric, 
                                              input_weights=input_weights)
        optimal_k, best_prediction = get_optimal_k(k_values, test_scores)
        print("When N equals", N, "the optimal k value is", optimal_k, "and the best test error rate  is", best_prediction)
        train_list.append(N)
        error.append(best_prediction)
        k_values.append(optimal_k)
        N+=10
        if(N > 210):
            boolean = False
#     print(len(train_list))
#     print(len(error), len(k_values))
    return train_list, error, k_values
```
