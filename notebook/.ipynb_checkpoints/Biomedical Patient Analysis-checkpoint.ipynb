## Writing functions to define the training and test sets

```python
# Function to train data
def get_train_data(data, n1, n2):
    class_0 = data[data.Class=='NO']
    class_1 = data[data.Class=='AB']
    x_train = pd.concat([class_0[:n1].iloc[:,:6], class_1[:n2].iloc[:,:6]])
    y_train = pd.concat([class_0[:n1].iloc[:,6:], class_1[:n2].iloc[:,6:]])
    x_train = np.array(x_train)
    y_train = np.array(y_train).ravel()
    return x_train, y_train
```

```python
def get_test_data(data, n1, n2):
    class_0 = data[data.Class=='NO']
    class_1 = data[data.Class=='AB']
    x_test = pd.concat([class_0[n1:].iloc[:,:6], class_1[n2:].iloc[:,:6]])
    y_test = pd.concat([class_0[n1:].iloc[:,6:], class_1[n2:].iloc[:,6:]])
    x_test = np.array(x_test)
    y_test = np.array(y_test).ravel()  
    return x_test, y_test
```

```python
def convert_numpy(x1,x2,y1,y2):
    x_train = np.array(x1)
    y_train = np.array(y1)
    x_test = np.array(x2)
    y_test = np.array(y2)
```

```python
#Function to covert categorical variables,"NO" and "AB", to numerical values 0 and 1.
def convert_classifier(input_val):
    encoder = preprocessing.LabelEncoder()
    output = encoder.fit_transform(input_val)
    return output
```

###  Selecting the first 70 rows of Class 0 and the first 140 rows of Class 1 as the training set and the rest of the data as the test set.

```python
x_test, y_test = get_test_data(data, 70,140)
```

```python
x_train, y_train = get_train_data(data, 70,140)
```

```python
y_train = convert_classifier((y_train))
y_test = convert_classifier((y_test))
```

### 1c.) Writing function for k-nearest neighbors with Euclidean metric

```python
#Function to build KNN model
def KNN_model(n, m, w, metric_params=None):
    model = KNeighborsClassifier(n_neighbors=n, metric = m, weights = w, metric_params=metric_params)
    return model
```

```python
#Function to retrieve k values and its train/test errors
def get_k_value_scores(x_train, y_train, x_test, y_test, increment, input_metric, input_weights, metric_params=None):
    k_values = []
    train_errors = []
    test_errors = []
    k = 1
    for x in range(len(x_train)):
        if(k < x): #number of neighbors has to be less than the size of training set
            model = KNN_model(k, input_metric, input_weights, metric_params=metric_params)
            model.fit(x_train, y_train)
            train_errors.append(1 - model.score(x_train, y_train))
            test_errors.append(1 - model.score(x_test,y_test))
            k_values.append(k)
            k += increment
    return k_values, train_errors, test_errors
```

### Getting train and test errors for all values of k where k ∈ {208, 205, 203...1}

```python
k_values, train_error, test_error = get_k_value_scores(x_train, y_train, x_test, y_test, 3, 'euclidean','uniform')
```

```python
#Function to plot line graph
def plot(x, y, xlabel, ylabel, invert, label):
    graph = plt.plot(x, y, label=label)
    if invert == True:
        plt.gca().invert_xaxis()
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.legend()
```

### Plotting train errors for values of k where k ∈ {208, 205, 203...1}

```python
plot(k_values, train_error, "K", "train_error", True, label="Train Error")
plot(k_values, test_error, "K", "train_error", True, label="Test Error")
plt.show()
```

```python
#Function to get the k value with the lowest error
def get_optimal_k(values, scores):
    best_error = min(scores)
    min_index = scores.index(best_error)
    min_index = k_values[min_index]
    return min_index, best_error
```

### For the training set, the best k-value of 1 yields an error rate of 0.0

```python
tr_optimal_k, tr_prediction = get_optimal_k(k_values, train_error)
print(tr_optimal_k)
print(tr_prediction)
```

### For the test set, the best k-value of 16 yields an error rate of 0.1

```python
optimal_k, prediction = get_optimal_k(k_values, test_error)
```

```python
print(prediction)
print(optimal_k)
```

```python
def get_data_accuracy(y_test, pred):
    true_negative, false_positive, false_negative, true_positive  = confusion_matrix(y_test,pred).ravel()
    precision = precision_score(y_test, pred)
    f1 = f1_score(y_test, pred)
    return true_negative, false_positive, false_negative, true_positive, precision, f1
```

```python
model = KNN_model(1, 'euclidean', 'uniform') #creating model with best k value
model.fit(x_train, y_train)
pred = model.predict(x_test)
true_negative, false_positive, false_negative, true_positive, precision, f1 = get_data_accuracy(y_test, pred)
```

```python
#Printing the values for the model
print("True Positive", true_positive)
print("True Negative", true_negative)
print("False Positive", false_positive)
print("False Negative", false_negative)
print("True Positive Rate", true_positive/(true_positive + false_positive))
print("True Negative Rate", true_negative/(true_negative+false_positive))
print("Precision:", precision)
print("F1 Score:", f1)
```

## Writing a function to get the best test error rates for all k values where k = {1,6,11,...196} for all N values where N = {10,20,30....210}

```python
# Getting values for N where N = {10,20,30....210}, test error rates for each value N, and the k values for each value N. 
def get_best_test_error(input_metric, input_weights):
    boolean = True
    N = 10
    train_list = []
    test_error = []
    k = []
    while boolean == True:
        x_train, y_train = get_train_data(data, int(round(N/3,0)), int(round((N - (N/3)),0)))
        x_test, y_test = get_test_data(data, int(round(N/3,0)), int(round((N - (N/3)),0)))
        y_train = convert_classifier(y_train)
        y_test = convert_classifier(y_test)
        k_values, train_scores, test_scores = get_k_value_scores(x_train, y_train, x_test, y_test,
                                              increment=5,
                                              input_metric=input_metric, 
                                              input_weights=input_weights)
        optimal_test_k, best_test_prediction = get_optimal_k(k_values, test_scores)
        train_list.append(N)
        test_error.append(best_test_prediction)
        k.append(optimal_test_k)
        N+=10
        if(N > 210):
            boolean = False

    
    return train_list, test_error, k
```

```python
n, test_e, k = get_best_test_error('euclidean', 'uniform')
```

```python
df_values = {"N": n, "Optimal K": k, "Test Error":test_e}
best_test_dataframe = pd.DataFrame(data=df_values)#, columns=["N", "Optimal K", "Best Test Error"])
```

## The best k value for the Minkowski model is 13

```python
mahalanobis_k_values, mahalanobis_train_errors, mahalanobis_test_errors = get_k_value_scores(x_train, y_train, x_test, y_test, 5, 'mahalanobis','uniform',
                                                                                             metric_params={'V': np.cov(x_train.T.astype(float)),
                                                                                                        'VI': None})
mah_optimal_k, mah_prediction = get_optimal_k(mahalanobis_k_values, mahalanobis_test_errors)
```

```python
euclidean_k_values, euc_train_errors, euc_test_errors = get_k_value_scores(x_train, y_train, x_test, y_test, 5, 'euclidean','distance')
euc_optimal_k, euc_prediction = get_optimal_k(euclidean_k_values, euc_test_errors)
```

```python
manhattan_k_values, manhattan_train_errors, manhattan_test_errors = get_k_value_scores(x_train, y_train, x_test, y_test, 5, 'manhattan','distance')
man_optimal_k, man_prediction = get_optimal_k(manhattan_k_values, manhattan_test_errors)
```

```python
cheb_k_values, cheb_train_errors, cheb_test_errors = get_k_value_scores(x_train, y_train, x_test, y_test, 5, 'chebyshev','distance')
cheb_optimal_k, cheb_prediction = get_optimal_k(cheb_k_values, cheb_test_errors)
```

```python
plot(mink_optimal_k, mink_prediction, "K values", "Test Errors", invert = False, label="Minkowski")
plot(mahalanobis_k_values, mahalanobis_test_errors, "K values", "Test Errors", invert = False, label="Mahalanobis")
plot(euclidean_k_values, euc_test_errors, "K Values", "Test Error", invert= False, label="Euclidean")
plot(manhattan_k_values, manhattan_test_errors, "K Values", "Test Error", invert= False, label="Manhattan")
plot(cheb_k_values, cheb_test_errors, "K Values", "Test Error", invert= False, label = "Chebyshev")
```
